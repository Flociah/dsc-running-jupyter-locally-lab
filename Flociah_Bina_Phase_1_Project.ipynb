# Data Visualization

Visualizing the number of movies per genre using a bar graph.

```python

genre_counts = df['genres'].value_counts()
plt.bar(genre_counts.index, genre_counts.values)
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.title('Number of Movies per Genre')
_ = plt.xticks(rotation=40, ha='right')
```


The number of movies under multiple genres were many followed by drama and comedy. The least genres produced are the sports, followed by romance and music genres.

**Visualizing the best language for the film**


```python
# Data visualization

x = df['original_language']
y= df['popularity']
fig, ax = plt.subplots(figsize=(12,10))
plt.bar(x, y, color='green')
plt.title('MOVIE PROFITABILITY BASED ON LANGUAGE AND POPULARITY')
plt.xlabel('original_language')
plt.ylabel('popularity')
plt.savefig('movie_profitability_based_on_language_and_popularity.png')
plt.show()
```

**Visualizing correlations**

```python
sns.heatmap(correlation, annot=True)
plt.rcParams['figure.figsize'] = (10, 10)  # Set the figure size
plt.title('Correlation Heatmap')
plt.savefig('correlation_heatmap.png')
plt.show()
```

**Visualizing central measures of tendency**

```python
# Box plot visuals of ROI by Genre
plt.figure(figsize=(20, 5)) # Increasing the size of the box plot.
sns.boxplot(x='genres', y='ROI', data=df)
plt.title('Box Plot of ROI by Genre')
plt.xlabel('Genre')
plt.ylabel('ROI')
plt.savefig('box_plot_roi_by_genre.png')
plt.show()
```

This is a box plot that shows how the genre is related to the Return Output Investment. As per the box plot, there are outliers and there is need to get rid of outliers to get accurate values

```python
# Boxplot without outliers
plt.figure(figsize=(20, 10))
sns.boxplot(x='genres', y='ROI', data=df, showfliers=False)
plt.title('Box Plot of ROI by Genre (Without Outliers)')
plt.xlabel('Genre')
plt.ylabel('ROI')
plt.savefig('box_plot_roi_by_genre_without_outliers.png')
plt.show()
```
